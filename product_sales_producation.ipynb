```python
# Enhanced Product Performance Analysis Project (Profile 7)

# Step 1: Import Libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
from sklearn.linear_model import LinearRegression

# Step 2: Load Dataset
# Sample dataset
data = {
    'Product': ['Apple','Banana','Orange','Grapes','Mango','Pineapple','Kiwi'],
    'Category': ['Fruit','Fruit','Fruit','Fruit','Fruit','Fruit','Fruit'],
    'Sales': [500, 300, 400, 200, 450, 350, 150],
    'Profit': [200, 100, 150, 80, 180, 130, 60],
    'Quantity_Sold': [50, 30, 40, 20, 45, 35, 15]
}
df = pd.DataFrame(data)
df.head()

# Step 3: Basic Data Exploration
df.info()
df.describe()
df.isnull().sum()

# Step 4: Visualize Product Sales
plt.figure(figsize=(10,5))
sns.barplot(x='Product', y='Sales', data=df, palette='viridis')
plt.title("Product Sales Performance")
plt.ylabel("Sales")
plt.xlabel("Product")
plt.xticks(rotation=45)
plt.show()

# Step 5: Visualize Profit per Product
plt.figure(figsize=(10,5))
sns.barplot(x='Product', y='Profit', data=df, palette='coolwarm')
plt.title("Profit per Product")
plt.ylabel("Profit")
plt.xlabel("Product")
plt.xticks(rotation=45)
plt.show()

# Step 6: Sales vs Profit Analysis
plt.figure(figsize=(8,6))
sns.scatterplot(x='Sales', y='Profit', hue='Product', size='Quantity_Sold', data=df, palette='Set2', s=200)
plt.title("Sales vs Profit Analysis")
plt.xlabel("Sales")
plt.ylabel("Profit")
plt.legend(bbox_to_anchor=(1.05, 1), loc=2)
plt.show()

# Step 7: Top Performing Products
# Top 3 products by sales
top_sales = df.sort_values('Sales', ascending=False).head(3)
print("Top 3 Products by Sales:\n", top_sales[['Product','Sales']])

# Top 3 products by profit
top_profit = df.sort_values('Profit', ascending=False).head(3)
print("\nTop 3 Products by Profit:\n", top_profit[['Product','Profit']])

# Step 8: Category Analysis
category_sales = df.groupby('Category')['Sales'].sum()
plt.figure(figsize=(6,6))
plt.pie(category_sales, labels=category_sales.index, autopct='%1.1f%%', startangle=140)
plt.title("Category-wise Sales Contribution")
plt.show()

# Step 9: Predictive Insight (Next Month Sales for Apple)
monthly_sales = [500, 520, 480, 550, 600, 580, 610]
months = np.arange(1, len(monthly_sales)+1).reshape(-1,1)
sales = np.array(monthly_sales)

model = LinearRegression()
model.fit(months, sales)
next_month = np.array([[len(monthly_sales)+1]])
forecast = model.predict(next_month)
print("Next Month Sales Forecast for Apple:", int(forecast[0]))

# Step 10: Insights & Recommendations
print("\nInsights:")
print("- Apple is the top-selling product with high profit.")
print("- Grapes and Kiwi have low sales; consider marketing or stock adjustments.")
print("- Profit per product shows where margins can be improved.")
print("- Forecast for Apple can help plan next month inventory.")
```
